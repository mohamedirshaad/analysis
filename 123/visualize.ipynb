```python
total_revenue =data["SALES"].sum()/1000000
```

```python
total_revenue= total_revenue.round(2)
print("Total Revenue: ",total_revenue," Millions")
```

```python
total_sales=data.groupby("TERRITORY")["SALES"].sum().sort_values(ascending=False)
```

```python
data_total_sales =total_sales.reset_index()
data_total_sales.columns = ["Region","Amount"]
```

```python
data_total_sales
```

```python
total_sales_by_country = data.groupby("COUNTRY")["ORDERNUMBER"].count().sort_values(ascending=False)
data_total_sales_by_count =total_sales_by_country.reset_index()
data_total_sales_by_count.columns ={"Country","No of Order"}
```

```python
data_total_sales_by_count
```

```python
most_bought_by_country =data.groupby("COUNTRY")["PRODUCTLINE"].agg(lambda x: x.value_counts().idxmax())
```

```python
data_most_bought_by_country = most_bought_by_country.reset_index()
data_most_bought_by_country.columns={"Country","Most frequently used"}
```

```python
data_most_bought_by_country
```
